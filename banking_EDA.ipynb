#  Exploratory Data Analysis of Banking Customer Behavior 

## Introduction 
This Exploratory Data Analysis (EDA) focuses on understanding customer demographics, financial behavior, risk exposure, and product usage in a banking environment. Using MySQL and Python (Pandas), the analysis aims to identify key patterns that support customer segmentation, risk assessment, and business decision-making.

```python
import mysql.connector
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

cnx=mysql.connector.connect(
    host="127.0.0.1",
    port=3306,
    user="root",
    password="root")
```

```python
query="SELECT * FROM banking_case.customer"
```

### Data Overview and Structure 

```python
df=pd.read_sql(query,cnx)
```

### Data Preprocessing and Quality Assessment

```python
df['Estimated Income'].value_counts()
```

```python
bins=[0,100000,300000, float('inf')]
labels= ['Low' , 'Med' , 'High']

df['Income Band'] = pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar',xlabel= 'Income Band')
```

Examine the distribution of unique categories in categorical columns

```python
categorical_cols=df[["BRId","GenderId","IAId","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

### Univariate Analysis of Customer Attributes

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
   
```

### Relationship Analysis Between Customer Attributes

```python
for i,predictor in enumerate(df[["BRId","GenderId","IAId","Amount of Credit Cards","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue="Nationality")
   
```

Histogram

```python
for col in categorical_cols:
    if col=="Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

### Correlation Analysis of Financial Variables

```python
numerical_cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account","Business Lending"]
```

Univariate Analysis and Visualization

```python
plt.figure(figsize=(10,15))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

### Heatmaps 

```python
numerical_cols=["Estimated Income","Superannuation Savings","Credit Card Balance","Bank Loans","Bank Deposits","Checking Accounts","Saving Accounts","Foreign Currency Account","Business Lending"]

correlation_matrix=df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

### Key Business Insights Derived from EDA

1.The strongest positive correlation occur among  "Bank Deposits" with "Checking Accounts","Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well

2.High-income customers are the primary contributors to deposit growth and are ideal candidates for premium banking services.

3.High credit card usage among lower-income customers highlights potential credit risk and the need for monitoring.

4.Loan exposure increases with customer risk level, making risk-based loan monitoring essential.

5.Customer loyalty directly correlates with higher product engagement and revenue contribution.

6.Property ownership can be used as a reliable factor for wealth segmentation.

## Conclusion
The EDA reveals clear relationships between income, product usage, risk, and customer loyalty. High-income and loyal customers form the bank’s most valuable segment, while high-risk customers with large loan exposure require careful monitoring. These insights can support better customer segmentation, targeted marketing, and improved risk management strategies.
